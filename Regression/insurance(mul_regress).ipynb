```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('insurance.csv')
data.head(10)
```

```python
data['sex'] = data['sex'].astype('category')
data['sex'] = data['sex'].cat.codes
data
```

```python
data['smoker'] = data['smoker'].astype('category')
data['smoker'] = data['smoker'].cat.codes

data['region'] = data['region'].astype('category')
data['region'] = data['region'].cat.codes
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.corr()
```

```python
data.isnull().sum()
```

```python
x = data.drop(columns = 'charges')
x
```

```python
y = data['charges']
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 24)
```

```python
from sklearn.linear_model import LinearRegression
lt = LinearRegression()
lt.fit(x_train , y_train)
```

```python
c = lt.intercept_
c
```

```python
m = lt.coef_
m
```

```python
y_pred_train = lt.predict(x_train)
y_pred_train.flatten()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train , y_pred_train)
```

```python
y_pred_test = lt.predict(x_test)
```

```python
y_pred_test
```

```python
plt.figure(figsize=(10,5))
plt.scatter(y_test , y_pred_test)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()] , 'k--', lw=2 , color = 'red')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title("Test Scatterplot")
```

```python
import numpy as np

coefficients = lt.coef_
feature_names = x.columns

feature_importance = pd.DataFrame({
    'Feature': feature_names,
    # 'Coefficient': coefficients,
    'Importance': np.abs(coefficients)
}).sort_values(by='Importance', ascending=False)

# print(feature_importance)
feature_importance['Importance'] = (feature_importance['Importance'] / feature_importance['Importance'].sum()) * 100
feature_importance
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(x = feature_importance['Feature'] , y = feature_importance['Importance'])
plt.xlabel('Feature Importance')
plt.ylabel('Percentage of Importance')
plt.xticks(rotation=45)
plt.title('Distribution of Feature Importance')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
mae = mean_absolute_error(y_test , y_pred_test) 
mse = mean_squared_error(y_test , y_pred_test) 
rmse = np.sqrt(mse)
mae , mse , rmse
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('star.csv')
data
```

```python
val = data.Beverage_category.value_counts().values.tolist()
ind = data.Beverage_category.value_counts().index.tolist()

plt.pie(val , labels = ind , autopct = '%1.1f%%')
plt.title("Beverage Category Available")
```

```python
bev_ind = data.Beverage.value_counts().index
bev_value = data.Beverage.value_counts().values

# plt.pie(bev_value , labels = bev_ind , autopct = '%1.1f%%')
plt.figure(figsize = (10,5))
sns.countplot(data, x= "Beverage_category")
plt.xticks(rotation = 65)
```

```python
plt.figure(figsize = (10,5))
sns.countplot(data, x= "Beverage_prep")
plt.xticks(rotation = 65)
plt.title("Preferred Beverage")
```

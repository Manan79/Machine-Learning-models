```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

```python
data = pd.read_csv('cardekho_dataset.csv')
data.drop(['Unnamed: 0' , 'km_driven' , 'mileage' , 'engine' , 'max_power'], axis=1, inplace=True)
```

```python
data
```

```python
OneHotEncoder = OneHotEncoder()

for column in data.columns:
    if data[column].dtype == 'object':
        
        data[column] = OneHotEncoder.fit_transform(data[column].values.reshape(-1, 1)).toarray()

        data[column] = data[column].astype(int)

data


```
